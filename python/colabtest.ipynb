```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
from pathlib import Path
from google.colab import drive

# 保存先のディレクトリ
save_dir = Path('expdata')
base_dir = Path('exp')

# 対象となるwavファイルをリスト化（ファイルが存在するか確認用）
wav_files = list(base_dir.glob('**/*.wav'))
print(f"見つかったwavファイルの数: {len(wav_files)}個")

for audio_dir in wav_files:
    print(f"処理中: {audio_dir}")
    
    y, sr = librosa.load(audio_dir)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    
    # 毎回新しいグラフのキャンバスを作成する
    plt.figure(figsize=(10, 4))
    
    librosa.display.specshow(mfccs, sr=sr, x_axis='time')
    plt.title('MFCC result')
    plt.xlabel('time (s)')
    plt.ylabel('MFCC Coefficients')
    plt.grid(True)
    plt.colorbar(label='Amplitude (dB)')
    plt.tight_layout()
    
    # グラフを表示
    plt.show()
    
    # メモリ解放のために図を閉じる（Colabでのループ処理では必須級）
    plt.close()
    
    genru = audio_dir.parent.name
    genru_save_dir = save_dir / genru
    genru_save_dir.mkdir(parents=True, exist_ok=True)
    save_path = genru_save_dir / f"{audio_dir.stem}_mfccs.npy"
    
    # 保存する
    np.save(save_path, mfccs)
    print(f"保存完了: {save_path}\n")
    drive.mount('/content/drive')
```
