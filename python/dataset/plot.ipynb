```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
from pathlib import Path

file_name = "5-263902-A-36.wav"
y, sr = librosa.load(file_name)

L = len(y)
t = np.arange(0,L) / sr

plt.plot(t,y)
plt.title('Audio waveform')
plt.xlabel('time(s)')
plt.ylabel('Amplitude')

plt.show()
```
